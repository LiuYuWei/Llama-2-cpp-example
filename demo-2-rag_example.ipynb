## Step 2: Transfer knowledge database to embedding vector

```python
from sentence_transformers import SentenceTransformer

# 載入模型
# 多語言模型庫：https://www.sbert.net/docs/pretrained_models.html?source=post_page-----db9c2a331b0f--------------------------------#multi-lingual-models
model = SentenceTransformer('paraphrase-multilingual-mpnet-base-v2')
```

```python
!wget https://raw.githubusercontent.com/LiuYuWei/Llama-2-cpp-example/main/aia-activity-rag.csv
```

```python
import pandas as pd

df = pd.read_csv("/content/aia-activity-rag.csv")
dataset_sentences = list(df['data'])
```

```python
# 計算數據集句子的嵌入向量
dataset_embeddings = model.encode(dataset_sentences)
```

## Step 3: Find similar sentence

```python
# 指定句子
query_sentence = "活動主辦單位是？"

# 你要找的前N個相似句子
top_n = 2
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# 計算查詢句子的嵌入向量
query_embedding = model.encode([query_sentence])

# 計算相似度
similarity_scores = cosine_similarity(query_embedding, dataset_embeddings)

# 獲取最相似句子的索引
top_n_indices = np.argsort(similarity_scores[0])[::-1][:top_n]

# 顯示最相似的句子和它們的相似度分數
for index in top_n_indices:
    print(f"句子: {dataset_sentences[index]}, 相似度分數: {similarity_scores[0][index]:.4f}")
```
